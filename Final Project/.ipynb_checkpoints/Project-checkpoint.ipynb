<h2>Remove data 'EDUACATION' and 'MARRIAGE' is  0</h2>

<p>we can see that the min value of EDUCATION and MARRIAGE is 0, but EDUCATION and MARRIAGE should be greater than 0, we check these rows first</p>

<h2> Add a new age column that divide the age (21 to 79) into 6 sections<h2>

<p>Write a function that returns a age range given age</p>

```python
def get_age_range(age):
    j = 20
    for i in range(6):
        if age < j+10:
            break
        j = j + 10
    return j
```

<p>add a new age column that divide the age (21 to 79) into 6 sections</p>

```python
df['age_range'] = df['AGE'].apply(lambda x:get_age_range(x))
```

```python
newage_group = df.groupby(['age_range','default.payment.next.month'])
ax= newage_group.size().unstack().plot(kind='bar')
fig = ax.get_figure()
filename = f'images/{"AGE"}.png'
fig.savefig(filename)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def scatterplot(x_data, y_data, x_label="", y_label="", title="", color = "r", yscale_log=False):

    # Create the plot object
    _, ax = plt.subplots()

    # Plot the data, set the size (s), color and transparency (alpha)
    # of the points
    ax.scatter(x_data, y_data, s = 10, color = color, alpha = 0.75)

    if yscale_log == True:
        ax.set_yscale('log')

    # Label the axes and provide a title
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```
