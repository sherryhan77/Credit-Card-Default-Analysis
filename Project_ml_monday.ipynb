<h2>Dataset Information</h2>
<p>This dataset contains information on default payments, demographic factors, credit data, history of payment, and bill statements of credit card clients in Taiwan from April 2005 to September 2005.<p>

<p>Content
There are 25 variables:
</p>
<p>
<p>ID: ID of each client
<p>LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit
<p>SEX: Gender (1=male, 2=female)
<p>EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
<p>MARRIAGE: Marital status (1=married, 2=single, 3=others)
<p>AGE: Age in years
<p>PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two <p>months, ... 8=payment delay for eight months, 9=payment delay for nine months and above)
<p>PAY_2: Repayment status in August, 2005 (scale same as above)
<p>PAY_3: Repayment status in July, 2005 (scale same as above)
<p>PAY_4: Repayment status in June, 2005 (scale same as above)
<p>PAY_5: Repayment status in May, 2005 (scale same as above)
<p>PAY_6: Repayment status in April, 2005 (scale same as above)
<p>BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)
<p>BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)
<p>BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
<p>BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
<p>BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
<p>BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)
<p>PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
<p>PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
<p>PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
<p>PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
<p>PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
<p>PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)
<p>default.payment.next.month: Default payment (1=yes, 0=no)

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()  #the accumulative weight

l= list(range(1,12)) + list(range(26,32))  #first 12 columns and column26-31
p=0.7
train_p=int(len(df)*p)

x_train = sc.fit_transform(df.iloc[0:train_p,l]) #a .b   #column 1,2
x_test=sc.fit_transform(df.iloc[train_p:,l ])
y_train=np.array(df.iloc[0:train_p,24])
y_test=np.array(df.iloc[train_p:,24])

# Support vector machine
from sklearn import svm
X = x_train
y = y_train
clf_svm = svm.SVC()
clf_svm.fit(X, y) 
y_svm = clf_svm.predict(x_test)

# neural network
from sklearn.neural_network import MLPClassifier
clf_nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
clf_nn.fit(X, y)
y_nn = clf_nn.predict(x_test)

# Decision trees
from sklearn import tree
clf_dt = tree.DecisionTreeClassifier()
clf_dt = clf_dt.fit(X, y)
y_dt = clf_dt.predict(x_test)

# Random forest
from sklearn.ensemble import RandomForestClassifier
clf_rf = RandomForestClassifier(n_estimators=10)
clf_rf = clf_rf.fit(X, y)
y_rf = clf_rf.predict(x_test)

# Naive bayes
from sklearn.naive_bayes import GaussianNB
bnb = GaussianNB()
y_bnb = bnb.fit(X, y).predict(x_test)
y_bnb

# Logistic regression
from sklearn import linear_model
clf_log = linear_model.LinearRegression()
clf_log.fit(X,y)

y_log = clf_log.predict(x_test)
def log_4(y_log):
    '''a function to convert result of logistic regression 
    to the value of 0 or 1, best threhold is 0.4'''
    for i in range(len(y_log)):
        if y_log[i] < 0.4:
            y_log[i]= 0
        else:
            y_log[i]= 1
    return y_log

# Knn
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
y_knn = neigh.predict(x_test)
```

```python
def confusion_matrix(y_log, y_test, threshold):
    #the threhold can help the prediction to be discrete
    if len(y_log) != len(y_test): return -1
    tp = 0.0
    fp = 0.0
    tn = 0.0
    fn = 0.0
    for i in range(len(y_test)):
        if y_test[i] ==1: #labels that are 1.0  (positive examples)
            if y_log[i] > threshold:
                tp += 1.0 #correctly predicted positive
            else:
                fn += 1.0 #incorrectly predicted negative
        else:              #labels that are 0.0 (negative examples)
            if y_log[i] <threshold:
                tn += 1.0 #correctly predicted negative
            else:
                fp += 1.0 #incorrectly predicted positive
    rtn = [tp, fp, tn, fn]

    return rtn
```

```python
def print_results(threshold):
    tp,fp,tn,fn = confusion_matrix(y_log,np.array(y_test),threshold)
    precision = tp/(tp+fp)
    tpr = tp/(tp+fn)
    tnr = tn/(tn+fp)
    fpr = fp/(fp+tn)
    f = precision*tpr/(precision+tpr)
    accuracy = (tp+tn)/(tp+tn+fp+fn)
    print("Accuracy:\t\t%1.2f how well the model has classified"%accuracy)
```

```python
def result(y_test, y_prediction):

    tp=0
    fp=0
    tn=0
    fn=0
    
    for i in range(len(y_test)):
        if y_test[i] == 1 and y_prediction[i]==1:
            tp+=1
        elif y_test[i]==0 and y_prediction[i]==1:
            fp+=1
        elif y_test[i]==0 and y_prediction[i]==0:
            tn+=1
        else:
            fn+=1
    tpr = tp/(tp+fn) #higher is better, if lower, can not detect out the default number
    tnr = tn/(tn+fp)  #higher is better
    fpr = fp/(fp+tn) #1-tnr
    precision = tp/(tp+fp)  # for postive or negative, right
    f = 2 * precision*tpr/(precision+tpr)
    accuracy = (tp+tn)/(tp+tn+fp+fn)  #total good number

    return ({'Precision':precision,'Recall':tpr, 'Specificity':tnr,'False Positive Rate':fpr,
             'f-score':f, 'Accuracy':accuracy,'tp':tp,'fp':fp,'tn':tn,'fn':fn})
```

```python
res_svm=result(y_test,y_svm)
res_nn=result(y_test,y_nn)
res_dt=result(y_test,y_dt)
res_rf=result(y_test,y_rf)
res_bnb=result(y_test,y_bnb)
res_knn = result(y_test,y_knn)
res_log = result(y_test,log_4(y_log))
```
