<h2>Dataset Information</h2>
<p>This dataset contains information on default payments, demographic factors, credit data, history of payment, and bill statements of credit card clients in Taiwan from April 2005 to September 2005.<p>

<p>Content
There are 25 variables:
</p>
<p>
<p>ID: ID of each client
<p>LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit
<p>SEX: Gender (1=male, 2=female)
<p>EDUCATION: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)
<p>MARRIAGE: Marital status (1=married, 2=single, 3=others)
<p>AGE: Age in years
<p>PAY_0: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two <p>months, ... 8=payment delay for eight months, 9=payment delay for nine months and above)
<p>PAY_2: Repayment status in August, 2005 (scale same as above)
<p>PAY_3: Repayment status in July, 2005 (scale same as above)
<p>PAY_4: Repayment status in June, 2005 (scale same as above)
<p>PAY_5: Repayment status in May, 2005 (scale same as above)
<p>PAY_6: Repayment status in April, 2005 (scale same as above)
<p>BILL_AMT1: Amount of bill statement in September, 2005 (NT dollar)
<p>BILL_AMT2: Amount of bill statement in August, 2005 (NT dollar)
<p>BILL_AMT3: Amount of bill statement in July, 2005 (NT dollar)
<p>BILL_AMT4: Amount of bill statement in June, 2005 (NT dollar)
<p>BILL_AMT5: Amount of bill statement in May, 2005 (NT dollar)
<p>BILL_AMT6: Amount of bill statement in April, 2005 (NT dollar)
<p>PAY_AMT1: Amount of previous payment in September, 2005 (NT dollar)
<p>PAY_AMT2: Amount of previous payment in August, 2005 (NT dollar)
<p>PAY_AMT3: Amount of previous payment in July, 2005 (NT dollar)
<p>PAY_AMT4: Amount of previous payment in June, 2005 (NT dollar)
<p>PAY_AMT5: Amount of previous payment in May, 2005 (NT dollar)
<p>PAY_AMT6: Amount of previous payment in April, 2005 (NT dollar)
<p>default.payment.next.month: Default payment (1=yes, 0=no)

<h2> Add a new age column that divide the age (21 to 79) into 6 sections<h2>

<p>Write a function that returns a age range given age</p>

```python
def get_age_range(age):
    j = 20
    for i in range(6):
        if age < j+10:
            break
        j = j + 10
    return j
```

<p>add a new age column that divide the age (21 to 79) into 6 sections</p>

```python
df['age_range'] = df['AGE'].apply(lambda x:get_age_range(x))
```
